```python
import pandas as pd
import os

# ============================================================
# 1. CARGAR ARCHIVO ORIGINAL
# ============================================================
df = pd.read_csv('Datos_Cundinamarca.csv')
print(f"Dataset original: {df.shape[0]} filas x {df.shape[1]} columnas")

# ============================================================
# 2. FILTRAR COLUMNAS NECESARIAS
# ============================================================
columnas_requeridas = [
    'PERIODO',
    'COLE_AREA_UBICACION',
    'COLE_BILINGUE',
    'COLE_GENERO',
    'COLE_JORNADA',
    'COLE_MCPIO_UBICACION',
    'ESTU_FECHANACIMIENTO',
    'ESTU_GENERO',
    'ESTU_PAIS_RESIDE',
    'FAMI_CUARTOSHOGAR',
    'FAMI_EDUCACIONMADRE',
    'FAMI_EDUCACIONPADRE',
    'FAMI_ESTRATOVIVIENDA',
    'FAMI_PERSONASHOGAR',
    'FAMI_TIENEAUTOMOVIL',
    'FAMI_TIENECOMPUTADOR',
    'FAMI_TIENEINTERNET',
    'FAMI_TIENELAVADORA',
    'PUNT_GLOBAL',
    'PUNT_MATEMATICAS',
    'FAMI_EDU_PADRES_SUM',
    'EDU_MAX_HOGAR',
    'INDICE_ECON'
]

columnas_existentes = [col for col in columnas_requeridas if col in df.columns]
columnas_faltantes  = [col for col in columnas_requeridas if col not in df.columns]

df_filtrado = df[columnas_existentes].copy()
print(f"\nDataset filtrado: {df_filtrado.shape[0]} filas x {df_filtrado.shape[1]} columnas")

# ============================================================
# 3. OPTIMIZAR TIPOS DE DATOS
# ============================================================
for col in df_filtrado.select_dtypes(include=['object']).columns:
    df_filtrado[col] = df_filtrado[col].astype('category')

for col in df_filtrado.select_dtypes(include=['float64']).columns:
    df_filtrado[col] = pd.to_numeric(df_filtrado[col], downcast='float')

for col in df_filtrado.select_dtypes(include=['int64']).columns:
    df_filtrado[col] = pd.to_numeric(df_filtrado[col], downcast='integer')

print(f"\nTamaño en memoria optimizado: {df_filtrado.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# ============================================================
# 4. GUARDAR ARCHIVO COMO df_filtrado.csv  ✅
# ============================================================
ruta_salida = 'df_filtrado.csv'
df_filtrado.to_csv(ruta_salida, index=False)

tamaño_mb = os.path.getsize(ruta_salida) / 1024**2
print(f"Archivo guardado: '{ruta_salida}' ({tamaño_mb:.2f} MB)")

# ============================================================
# 5. VERIFICAR COLUMNAS
# ============================================================
print("\n" + "="*50)
print("VERIFICACIÓN DE COLUMNAS")
print("="*50)

presentes = [col for col in columnas_requeridas if col in df_filtrado.columns]
faltantes  = [col for col in columnas_requeridas if col not in df_filtrado.columns]

print(f"\n✅ Columnas presentes ({len(presentes)}/{len(columnas_requeridas)}):")
for col in presentes:
    print(f"   - {col:30s} {str(df_filtrado[col].dtype)}")

if faltantes:
    print(f"\n❌ Columnas NO encontradas en el dataset original ({len(faltantes)}):")
    for col in faltantes:
        print(f"   - {col}")
else:
    print("\n🎉 Todas las columnas requeridas están presentes")

print("\n", df_filtrado.head())

# Reemplaza la línea de guardado por esta:
ruta_salida = 'df_filtrado.csv.zip'
df_filtrado.to_csv(ruta_salida, index=False, compression='zip')

tamaño_mb = os.path.getsize(ruta_salida) / 1024**2
print(f"Archivo guardado: '{ruta_salida}' ({tamaño_mb:.2f} MB)")
```

```python
#librerias necesarias
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
```

## Redes Neuronales

```python
#datos para el modelo
df_modelo = pd.read_csv("df_modelo_preg1.csv")
df_modelo.head()
```

```python
#caracteristicas
X = df_modelo[['COLE_AREA_UBICACION_NUM',
               'COLE_BILINGUE_NUM',
               'COLE_GENERO_NUM',
               'COLE_JORNADA_NUM']].values

#variable predictiva del modelo 
y = df_modelo['NIVEL_MATEMATICAS'].values

# target a índices 0, 1, 2
y_idx = y - 1

#target de la red
y_cat = to_categorical(y_idx, num_classes=3)
```

```python
#80% train, 20% test

X_train, X_test, y_train, y_test, y_idx_train, y_idx_test = train_test_split(
    X, y_cat, y_idx,
    test_size=0.2,
    random_state=42,
    stratify=y_idx
)

#escalar features
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)
```

```python
#verificar
print(f"X_train: {X_train_sc.shape} | X_test: {X_test_sc.shape}")
print(f"y_train: {y_train.shape}   | y_test: {y_test.shape}")
print(f"\nDistribución en test:")
unique, counts = np.unique(y_idx_test, return_counts=True)
for u, c in zip(unique, counts):
    nombre = {0:'Bajo', 1:'Medio', 2:'Alto'}[u]
    print(f"  {nombre}: {c} ({c/len(y_idx_test)*100:.1f}%)")
```
